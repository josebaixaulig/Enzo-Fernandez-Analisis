```python
import pandas as pd 
sofascore_mid=pd.read_csv('data/sofascore_premier_defensive_stats_midfielders.csv')
fbref_mid=pd.read_csv('data/premier_midfielders_atacking_stats.csv')
```

```python
sofascore_mid = sofascore_mid.drop_duplicates()
```

Cambiamos nombres para que sean iguales

```python
from difflib import get_close_matches

# Obtener los nombres que no coinciden
missing_names = set(sofascore_mid['Name']) - set(fbref_mid['Player'])
missing_names_list = set(fbref_mid['Player']) - set(sofascore_mid['Name'])

# Encontrar los nombres más parecidos
matches = {}
for name in missing_names:
    close_match = get_close_matches(name, missing_names_list, n=1, cutoff=0.8)  # Ajusta 'cutoff' si es necesario
    matches[name] = close_match[0] if close_match else None

# Cambiar los nombres en sofascore_mid
sofascore_mid['Name'] = sofascore_mid['Name'].replace(matches)

# Imprimir los cambios realizados
for original, match in matches.items():
    if match:
        print(f"Cambiado: {original} -> {match}")
    else:
        print(f"{original} -> Sin coincidencia")
```

```python
fbref_mid
```

```python
midfielders_stats=pd.merge(fbref_mid,sofascore_mid,left_on='Player',right_on='Name',how='inner')
midfielders_stats.to_csv('midfielders_stats.csv',index=False)
```

```python
import plotly.express as px
import pandas as pd

# Crear un DataFrame de ejemplo
data = {'Metric': ['Passing', 'Dribbling', 'Shooting', 'Defending', 'Physicality'],
        'Enzo Fernandez': [85, 70, 80, 65, 90],
        'Martin Odegard ': [75, 60, 85, 70, 80]}
df = pd.DataFrame(data)

# Transformar el DataFrame para Plotly
df_melted = df.melt(id_vars=['Metric'], var_name='Player', value_name='Score')

# Crear el radar chart
fig = px.line_polar(df_melted, r='Score', theta='Metric', color='Player', line_close=True, 
                    title="Radar Chart: Player Comparison")
fig.show()
```
